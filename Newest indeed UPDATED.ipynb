```python
job_title= 'Data Analyst' #you can change this to any job position you are searcing for 
location = 'New York' # You can change this to any location you are searching in 
pageNum = 5 #how many pages you want to scrape 
```

```python
from bs4 import BeautifulSoup
import re
import time
import requests
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
import pandas as pd
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
def check_exists_by_xpath(xpath):
    try:
        browser.find_element_by_xpath(xpath)
    except NoSuchElementException:
        return False
    return True


def pageLink(job_title, location):
    test = 'https://www.indeed.com/jobs?q={}&l={}'
    url = test.format(job_title,location)
    return url
```

```python
pageLink = pageLink(job_title , location)
path = "/Users/katia/Desktop/chromedriver"
browser = webdriver.Chrome(path)
browser.get(pageLink)
page_source = browser.page_source
soup = BeautifulSoup(page_source, 'lxml')
time.sleep(2)
```

```python
dataframe = pd.DataFrame(columns = ['Position', 'Location', 'Company', 'Rating', 'Salary', 'Description'])
page_source = browser.page_source
soup = BeautifulSoup(page_source, 'lxml')

jobs = soup.find_all('div', 'job_seen_beacon')

for i in range(0,pageNum ):
    for job in jobs:
    

        try:
            title = job.find('h2' , class_ = 'jobTitle').text.strip('new')
           
        except:
            title = 'NA'
        try:
            location = job.find('div' ,class_ = "companyLocation").text
        
        except:
            location = 'NA'
    
        try:
            company = job.find('span' ,class_ = 'companyName').text.replace('\n', '').strip()
        
        except:
            company = "NA"
        
        
        try:
            rating = job.find('span' , class_ = 'ratingNumber').text.strip()
        except:
            rating = 'NA'
            
        try:
            salary = job.find('div', class_ = 'attribute_snippet' ).text.strip()
        except:
            salary = 'NA'
        
        #summary = browser.find_element_by_xpath('/html/body/table[2]/tbody/tr/td/table/tbody/tr/td[1]/div[5]/div/a[1]/div[1]/div/div[1]/div/table[1]/tbody/tr/td/div[1]/h2/span')
        #summary.click()
        
        
        
        
    
        try:   
            x = job.find('div' , class_ = 'more_links')
            y = str(x)
            k = y.split("jk=")
            ec = k[1][0:16]
            element = '//*[@id="job_'+ec+'"]/div[1]/div/div[1]/div/table[1]/tbody/tr/td/div[1]/h2/span'
            browser.find_element_by_xpath(element).click()
            time.sleep(2)
            
        except:
            description = 'NA'
            
        try:
            description = browser.find_element_by_id('jobDescriptionText').text
                  
        except:
            description = 'NA'
            
        dataframe = dataframe.append({'Position': title, 'Location': location, 'Company': company, 'Rating':rating , 'Salary':salary, 'Description': description}, ignore_index = True)
        
        
        ## The below code will click out of the pop up if it comes up 
        try:
            popoverbutton = browser.find_element_by_xpath('/html/body/div[5]/div[1]/button')
            popoverbutton.click()
        
        except:
            continue
            
        time.sleep(2)
            
   ## Checking if next page button exists, if it does, click it
    try:
        next_page = browser.find_element_by_xpath('//a[@aria-label={}]//span[@class="pn"]'.format(i+2))
        next_page.click()

    except:
        next_page = browser.find_element_by_xpath('//a[@aria-label="Next"]//span[@class="np"]')
        next_page.click()
    
        
    
    print("Page: {}".format(str(i+2)))
    
```
